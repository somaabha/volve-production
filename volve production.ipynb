```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("C:/Users/somaa/Downloads/Volve production data.xlsx")
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,6))
plt.title("Oil production for all wells")
plt.xlabel("Production year")
plt.ylabel("Oil volume in scm")
sns.lineplot(x="DATEPRD", y="BORE_OIL_VOL", data=df, hue ="NPD_WELL_BORE_NAME")
```

```python
plt.figure(figsize=(15,6))
plt.title("Gas production for all wells")
plt.xlabel("Production year")
plt.ylabel("Gas volume in scm")
sns.set_palette("Set2")
sns.lineplot(x="DATEPRD", y="BORE_GAS_VOL", data=df, hue ="NPD_WELL_BORE_NAME")
```

```python
plt.figure(figsize=(15,6))
plt.title("Water production for all wells")
plt.xlabel("Production year")
plt.ylabel("Water volume in scm")
sns.set_palette("Set2")
sns.lineplot(x="DATEPRD", y="BORE_WAT_VOL", data=df, hue ="NPD_WELL_BORE_NAME")
```

```python
# contribution of each well in oil production 
oilprod=df.groupby("NPD_WELL_BORE_NAME")["BORE_OIL_VOL"].sum()
oilprod
```

```python
# contribution of each well in gas production 
gasprod=df.groupby("NPD_WELL_BORE_NAME")["BORE_GAS_VOL"].sum()
gasprod
```

```python
# contribution of each well in water production 
waterprod=df.groupby("NPD_WELL_BORE_NAME")["BORE_WAT_VOL"].sum()
waterprod
```
